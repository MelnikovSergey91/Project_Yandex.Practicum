## 2. Подготовка данных

Переработаем данные таблицы по выводам, сделанным в первом шаге.

 ### 2.1. Изменение регистра написания названий столбцов и добавление нового столбца all_sales

Для этого воспользуемся методом **str.lower()** и сразу проверим

```python
data.columns = data.columns.str.lower()
data['all_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales']
data.head(5)
```

### 2.1. Вывод

Все названия столбцов теперь написаны в нижнем регистре. Добавили столбец **all_sales**, в нем отображаются суммарные продажи по всем регионам.

### 2.2. Работа с пропусками и типом данных

**Critic_Score - оценка критиков** меняем все пустые значения на **-1** и проведем проверку

```python
data['critic_score'] = data['critic_score'].fillna(-1)
data['critic_score'].isnull().sum()
```

**Rating - рейтинг от организации ESRB** меняем все пустые значения на **"NR"** и проведем проверку

```python
data['rating'] = data['rating'].replace(np.NaN, 'NR')
data['rating'].isnull().sum()
```

**User_Score - оценка пользователей** чтобы понять на что поменять, сначала выясним, почему по данному столбцу тип данных **object**, для этого посмотрим все уникальные значения по столбцу методом **unique**

```python
data['user_score'].unique()
```

А вот и причина - **tbd**, выясним сколько строк с таким значением

```python
len(data[data['user_score'] == 'tbd'])
```

2424 строки имеют данное значение. Покапавшись в интернете можно легко понять что **tbd** это абревиатура и расшифровывается очень просто - To Be Determined, то есть "Будет определено". Предполагаю, что по этим играм, просто не набралось достаточного колличества оценок, чтобы они имели какой то пользовательский рейтинг. Заменить данные значения на среднее или медиану будет не корректным, поставить значение **0** тоже нельзя, тк в этом столбце есть игры с "нулевым" рейтингом, поэтому присвоим этим строка новое значение **-1**. Это позволит не перемешать данные этих строк с реальными значениями рейтинга и при этом мы сможем продолжить анализ тих данных. Также сразу поменяем тип данных для данного столбца на **float**.

```python
data['user_score'] = data['user_score'].replace('tbd', np.NaN)
data['user_score'] = data['user_score'].fillna(-1)
data['user_score'] = data['user_score'].astype('float')
data['user_score'].isnull().sum()
```

**Name - название игры и Genre - жанр игры** посмотрим совпадают ли строки, где пропущено значение в данных столбцах

```python
data.query('name.isnull() and genre.isnull()')
```

Как я и предполагал, строки совпадают, заменить их на что-то стоящее не возможно, удалим их и сделаем проверку.

```python
data = data.drop(index = data.query('name.isnull() and genre.isnull()').index)
data = data.reset_index(drop=True)
display(data['genre'].isnull().sum())
display(data['name'].isnull().sum())
```

### 2.2. Вывод

Избавились от пустых значений в столбцах:

- **Critic_Score - оценка критиков** - заменили их на *0*

### 3.5. Влияние на продажи внутри одной популярной платформы отзывы пользователей и критиков

Для примера возьмем платформу **PS4**, тк она долго и давно на рынке продаж, по ней много данных и на конец 2016 года она еще используется пользователями.

Сначала построим диаграмму рассеивания зависимости продаж от отзывов критиков

Построим диаграммы рассеивания зависимости продаж от отзывов критиков и отзывов пользователей

```python
def score(score, title, xlabel):
    sales_by_critic = data_period[(data_period['platform'] == 'PS4') & (data_period[score] >= 0)]
    plt.figure(figsize=(15,7))
    sns.set()
    sns.scatterplot(x=score ,y='all_sales',data=sales_by_critic)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Продажи(млн. копий)')
    plt.show()
score('critic_score', 'Зависимость продаж от отзывов критиков', 'Отзывы критиков (оценка от 1 до 100)')
score('user_score', 'Зависимость продаж от отзывов пользователей', 'Отзывы пользователей (оценка от 1 до 10)')
```

- Из графика оценки критиков видно, что чем выше рейтинг критиков, тем дороже игра. Все вполне логично, чем лучше сделана игра, тем лучше оценки критиков и тем выше продажи по этой игре.
- Исходя из графика оценки пользователей видим, по аналогии с оценками критиков, что чем выше оценка, тем больше продаж имеет игра. Однако, мы также можем заметить выбросы, где оценка достаточно посредственная, но при этом игра достаточно хорошо продавалась. Предполагаю что это связано с маркетингом и возмонжо "именем игры", тоесть рекламная компания по продажам данной игры была очень хорошая или эта игра была из "серии игр" которые ранее уже сделали себя "имя", , что привело к большим продажам но при этом сама игра пользователей не впечатлила что и привело к низким оценкам. 

```python
sales_by_critic_users =(
    data_period.query('platform == "PS4" and user_score >= 0 and critic_score >= 0')
)

#Расчет корреляции
cor1 = (sales_by_critic_users
    [['user_score', 'critic_score', 'all_sales']].corr()
    [['all_sales']].sort_values(by='all_sales', ascending=False)
)

plt.figure(figsize=(3, 5))
sns.set()
sns.heatmap(cor1, annot=True, fmt='.2f')
plt.yticks(rotation=0)
plt.title('Зависимость между отзывами и продажами')
plt.show()
```

Отзывов критиков на продажи по игре влияют, больше чем у отзывов пользователей. Предполагаю что это связано с тем что критикам, сами пользователи доверяют больше, чем обычным пользователям, как они сами, отсюда и получается, что чем выше оценка критиков, тем больше продаж по игре, и очень мало выбросов, тоесть критики не оценивают рекламные компнаии по продаже игры или "имя", которое было заработано ранее, они смотрят на саму игру и дают ей объективную оценку.

### 3.5. Вывод

На примере платформы **PS4** выяснили, что есть прямая зависимость, чем больше оценка критиков/пользователей тем больше продаж имеет игра. Однако в оценками пользователей есть еще "выбросы", где стоимость игры достаточно большая, при достаточно непосредсвенных оценка, предполагаю что это связано с маркетинговой компание или "именем игры", что дает ей большие продажи, но после того как пользователи в нее поиграли, оценили ее достаточно низко. Оценки же критиков, данным влияниям не подвержны и имеют большее влияние на конечные продажи игры.

### 3.6.  Соотношение выводов с продажами игр на других платформах

Построим графики корреляции по глобальным продажам игр в разбивке по самым топовым платформам, для этого напишем функцию для построения графиков

```python
def heatmap (platform_name, title):
    #Выделяем итоговые списки
    sales_by_critic_users_top6 = data_period[(data_period['platform'] == platform_name) &
        (data_period['user_score'] >= 0) & (data_period['critic_score'] >= 0)]
    #Считаем корреляцию
    cor2 = (sales_by_critic_users_top6
        [['user_score', 'critic_score', 'all_sales']].corr()
        [['all_sales']].sort_values(by='all_sales', ascending=False)
           )
    plt.figure(figsize=(5,1))
    sns.set()
    sns.heatmap(data=cor2, annot=True, fmt='.2f')
    plt.yticks(rotation=0)
    plt.title(title)
```

Построим графики для топовых платформ, за исключением **PS** и **PS2**, тк к 2011 году они полность себя исчерпали, и добавим еще платформы **PS4, 3DS и XOne**, тк эти платформы вышли не давно, но уже имеют хорошие продажи.

```python
heatmap('X360', 'Зависимость между отзывами и продажами для X360')
heatmap('Wii', 'Зависимость между отзывами и продажами для Wii')
heatmap('PS3', 'Зависимость между отзывами и продажами для PS3')
heatmap('DS', 'Зависимость между отзывами и продажами для DS')
heatmap('3DS', 'Зависимость между отзывами и продажами для 3DS')
heatmap('XOne', 'Зависимость между отзывами и продажами для XOne')
heatmap('PS4', 'Зависимость между отзывами и продажами для PS4')
```

### 3.6. Вывод 

Для платформ **X360, PS3, PS4, и XOne** оценка критиков очень важна, тк она сильно влияет на продажи, на платформе **DS** чуть меньше. ### 4.3. Влияние рейтинга ESRB на продажи в отдельном регионе

Добавив в таблицу новый столбец с расшифровкой рейтинга, для удобства понимания

```python
def decryption(row):
    if row['rating'] == 'NR':
        return 'нет рейтинга'
    elif row['rating'] == 'M':
        return 'старше 18 лет'
    elif row['rating'] == 'T':
        return '>= 13 лет и старше'
    elif row['rating'] == 'E10+':
        return 'для всех от 10 лет и старше' 
    elif row['rating'] == 'AO':
        return 'Только для взрослых'
    elif row['rating'] == 'EC':
        return 'для детей от 3 лет'
    elif row['rating'] == 'RP':
        return 'рейтинг ожидается'
    elif row['rating'] == 'E' or row['rating'] == 'K-A':
        return 'для всех'
data_period['esbr_decryption'] = data_period.apply(decryption, axis=1)
data_period.head(10)
```

Определим самые популярные рейтинги в заданных регионах, для этого напишем функцию, которая строит графики

```python
def by_esbr(index, region_sales, title, xlabel):
    not_nr = data_period.query('rating != "NR"')
    #Выделяем итоговые списки
    not_nr_table = (not_nr.pivot_table
        (index =index, values = region_sales, aggfunc='sum')
        .sort_values(by=region_sales, ascending=False)
                   )
    #Строим график
    plt.figure(figsize=(15,6))
    sns.set()
    sns.barplot(x=not_nr_table.index, y=region_sales, data=not_nr_table)
    plt.xticks(rotation=90)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Продажи (мнл. копий)')
    plt.show()
```

Посмтроим график по продажам игр в Северной Америке в зависимости от рейтинга ESRB

## 5. Проверка гипотез

- Средние пользовательские рейтинги платформ **XOne** и **PC** одинаковые;
- Средние пользовательские рейтинги жанров **Action** (англ. «действие», экшен-игры) и **Sports** (англ. «спортивные соревнования») разные.

Для проверки гипотез будем пользоваться **t-критерием Стьюдента**, тк выборки между собой независимы.
Пороговое значение **alpha** определим как равное **0.05**.

### 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформируем **нулевую гипотезу Н0** и **альтернативную Н1**:

- Н0 - средние пользовательские рейтинги платформ **XOne** и **PC** одинаковые.
- Н1 - средние пользовательские рейтинги платформ **XOne** и **PC** отличаются между собой.

Сохраним необходимые нам выборки для проверки гипотез.

```python
xone_users = data_period[(data_period['platform'] == 'XOne') & (data_period['user_score'] >= 0)]['user_score']
pc_users = data_period[(data_period['platform'] == 'PC') & (data_period['user_score'] >= 0)]['user_score']
```

Чтобы понять, какой параметр применить к **equal_var**, проверим равенство дисперсий выборок.

```python
stat, p = levene(xone_users, pc_users)
print('p-значение:', p)
```

Значение p < 0.05 говорит о том, что выборки не имеют однородные дисперсии, значит **equal_var** присвоим значение **False**. 

Проверим гипотезу о том, что средние пользовательские рейтинги платформ XOne и PC одинаковые.

```python
alpha = 0.05
results = st.ttest_ind(xone_users, pc_users, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### 5.1. Вывод

Нулевую гипотезу отвергнуть не получается. Верно утверждение, средние пользовательские рейтинги платформ **XOne и PC** одинаковые.

Значение p < 0.05 говорит о том, что выборки не имеют однородные дисперсии, значит **equal_var** присвоим значение **False**.

Проверим гипотезу о том, что средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.

```python
alpha = 0.05
results = st.ttest_ind(action_users, sports_users, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### 5.2. Вывод

Отвергаем нулевую гипотезу. Верно утверждение, средние пользовательские рейтинги жанров **Action и Sports** отличаются между собой.
